```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1], [2], [3], [4], [5]], dtype=float)
y = np.array([[2], [4], [6], [8], [10]], dtype=float)
```

```python
np.random.seed(42)

W = np.random.randn(1, 1)
b = np.random.randn(1)

learning_rate = 0.01
epochs = 10
losses = []
```

```python
for epoch in range(epochs):

    # Forward Pass
    y_hat = np.dot(X, W) + b

    # Compute Loss (Mean Squared Error)
    loss = np.mean((y - y_hat) ** 2)
    losses.append(loss)

    # Backpropagation
    dW = -2 * np.dot(X.T, (y - y_hat)) / len(X)
    db = -2 * np.mean(y - y_hat)

    # Update Parameters
    W -= learning_rate * dW
    b -= learning_rate * db
```

```python
#  if epoch % 100 == 0:
#         print(f"Epoch {epoch}, Loss: {loss:.6f}")
```

```python
print("\nLearned Weight (W):", W)
print("Learned Bias (b):", b)
```

```python
test_input = np.array([[7]], dtype=float)
prediction = np.dot(test_input, W) + b
```

```python
print("\nPrediction for 7:", prediction)
```

```python
plt.figure()
plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss vs Epochs")
plt.show()
```
